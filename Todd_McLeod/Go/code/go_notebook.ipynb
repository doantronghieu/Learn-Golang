### PASS BY VALUE

```go
import (
	"fmt"
	"sort"
)

func medianOne(x []float64) (float64) {
	sort.Float64s(x)

	i := len(x) / 2
	if len(x) % 2 == 1 {
		return x[i / 2]
	}
	return (x[i - 1] + x[i]) / 2
}

n := []float64{3, 1, 4, 2}

fmt.Println(n)
fmt.Println(medianOne(n))
fmt.Println(n)

```

```go
import (
	"fmt"
	"sort"
)

func medianTwo(x []float64) (float64) {
	n := make([]float64, len(x))
	copy(n, x)

	sort.Float64s(n)

	i := len(n) / 2
	if len(n) % 2 == 1 {
		return n[i / 2]
	}
	return (n[i - 1] + n[i]) / 2
}

n := []float64{3, 1, 4, 2}

fmt.Println(n)
fmt.Println(medianTwo(n))
fmt.Println(n)

```

```go
package main

import (
	"fmt"
)

type person struct {
	firstName string
	lastName string
	favIC []string
}

func main() {
	p1 := person{
		firstName: "James",
		lastName: "Bond",
		favIC: []string{"chocolate", "banana", "mango"},
	}

	p2 := person{
		firstName: "Jenny",
		lastName: "MonneyPenny",
		favIC: []string{"strawberry", "chocolate"},
	}

	m := map[string]person{
		p1.lastName: p1,
		p2.lastName: p2,
	}

	fmt.Println(p1)
	fmt.Println(p2)
	fmt.Println()
	
	fmt.Println(p1.favIC)
	fmt.Println(p2.favIC)
	fmt.Println()

	for _, v := range p1.favIC {
		fmt.Println(p1.firstName, "favorite is", v)
	}
	for _, v := range p2.favIC {
		fmt.Println(p2.firstName, "favorite is", v)
	}
	fmt.Println()

	for _, v := range m {
		fmt.Println(v)
		for _, v2 := range v.favIC {
			fmt.Printf("%v ", v2)
		}
		fmt.Println()
	}
}

main()
```

```go
package main

import (
	"fmt"
	"math/rand"
)

func main() {
	x := rand.Intn(10)
	y := rand.Intn(10)

	fmt.Printf("x: %v, y: %v\n", x, y)

	if (x < 4) && (y < 4) {
		fmt.Println("Both are less than four")
	} else if (x > 6) && (y > 6) {
		fmt.Println("Both are greater than six")
	} else if (x >= 4) && (x <= 6) {
		fmt.Println("x is from 4 to 6 inclusive")
	} else if (y != 5) {
		fmt.Println("y is not five")
	} else {
		fmt.Println("None of previous are met.")
	}

	switch {
	case (x < 4) && (y < 4):
		fmt.Println("Both are less than four")
	case (x > 6) && (y > 6):
		fmt.Println("Both are greater than six")
	case (x >= 4) && (x <= 6):
		fmt.Println("x is from 4 to 6 inclusive")
	case (y != 5):
		fmt.Println("y is not five")
	default:
		fmt.Println("None of previous are met.")		
	}
}

main()
```

```go
package main

import (
	"fmt"
	"time"
	"math/rand"
)

func main() {
	ch1 := make(chan int)
	ch2 := make(chan int)

	d1 := time.Duration(rand.Int63n(250))
	d2 := time.Duration(rand.Int63n(250))

	go func() {
		time.Sleep(d1 * time.Millisecond)
		ch1 <- 41
	}()

	go func() {
		time.Sleep(d2 * time.Millisecond)
		ch2 <- 42
	}()

	select {
	case v1 := <- ch1:
		fmt.Println("value from channel 1: ", v1)
	case v2 := <- ch2:
		fmt.Println("value from channel 2: ", v2)
	}
}

main()
```

### INTERFACE, POLYMORPHISM 

```go
package main

import (
	"fmt"
	"math"
)

type Square struct {
	length float64
	width float64
}

type Circle struct {
	radius float64
}

func (s Square) Area() (float64) {
	return s.length*s.width
}

func (c Circle) Area() (float64) {
	return math.Pi*math.Pow(c.radius, 2)
}

type Shape interface {
	Area() (float64)
}

func Info(s Shape) (float64) {
	return s.Area()
}

func main() {
	c1 := Circle{
		radius: 4,
	}

	s1 := Square{
		length: 5,
		width: 8,
	}

	fmt.Println(Info(c1))
	fmt.Println(Info(s1))
}

main()
```

```go
package main

import (
	"fmt"
	"log"
)

type User struct {
	ID int
	FirstName string
}

type MockDataStore struct {
	Users map[int]User
}

func (md MockDataStore) GetUser(id int) (User, error) {
	user, ok := md.Users[id]
	if !ok {
		return User{}, fmt.Errorf("User %v not found", id)
	}
	return user, nil
}

func (md MockDataStore) SaveUser(u User) (error) {
	_, ok := md.Users[u.ID]
	if ok {
		return fmt.Errorf("User %v already exists", u.ID)
	}
	md.Users[u.ID] = u
	return nil 
}

type Datastore interface {
	GetUser(id int) (User, error)
	SaveUser(u User) (error) 
}

type Service struct {
	ds Datastore
}

func (s Service) GetUser(id int) (User, error) {
	return s.ds.GetUser(id)
}

func (s Service) SaveUser(u User) (error) {
	return s.ds.SaveUser(u)
}

func main() {
	db := MockDataStore{
		Users: make(map[int]User),
	}

	service := Service{
		ds: db,
	}

	u1 := User{
		ID: 1,
		FirstName: "James",
	}

	err := service.SaveUser(u1)
	if (err != nil) {
		log.Fatalf("Error %s", err)
	}

	u1Returned, err := service.GetUser(u1.ID)
	if (err != nil) {
		log.Fatalf("Error %s", err)
	}

	fmt.Println(u1)
	fmt.Println(u1Returned)
}

main()
```

```go
import (
	"fmt"
	"sort"
)

type Person struct {
	FirstName string
	Age int
}

func (p Person) String() string {
	return fmt.Sprintf("%s: %d", p.FirstName, p.Age)
}
// ----------------------------------------------------------------------------
type ByAge []Person

func (a ByAge) Len() int {
	return len(a)
}

func (a ByAge) Swap(i, j int) {
	a[i], a[j] = a[j], a[i]
}

func (a ByAge) Less(i, j int) bool {
	return a[i].Age < a[j].Age
}
// ----------------------------------------------------------------------------
type ByName []Person

func (bn ByName) Len() int {
	return len(bn)
}

func (bn ByName) Swap(i, j int) {
	bn[i], bn[j] = bn[j], bn[i]
}

func (bn ByName) Less(i, j int) bool {
	return bn[i].FirstName < bn[j].FirstName
}
// ----------------------------------------------------------------------------

func main() {
	p1 := Person{"James", 32}
	p2 := Person{"MonneyPenny", 27}
	p3 := Person{"Q", 64}
	p4 := Person{"M", 56}

	people := []Person{p1, p2, p3, p4}
	fmt.Println(people)

	sort.Sort(ByAge(people))
	fmt.Println(people)

	sort.Sort(ByName(people))
	fmt.Println(people)
}
```

## MUTEX

```go
import (
	"fmt"
	"runtime"
	"sync"
	"time"
)

func main() {
	fmt.Println("CPUs: ", runtime.NumCPU())
	fmt.Println("Goroutines: ", runtime.NumGoroutine())

	counter := 0

	var wg sync.WaitGroup
	var mu sync.Mutex

	const gs = 100
	wg.Add(gs)

	fmt.Println()
	fmt.Printf("Goroutines: %v ", runtime.NumGoroutine())
	for i := 0; i < gs; i++ {
		go func() {
			mu.Lock()

			v := counter
			runtime.Gosched()
			v++
			counter = v
			
			mu.Unlock()
			wg.Done()
		}()
		fmt.Printf("%v ", runtime.NumGoroutine())
	}
	fmt.Println()

	wg.Wait()

	fmt.Println("CPUs: ", runtime.NumCPU())
	fmt.Println("Goroutines: ", runtime.NumGoroutine())
	fmt.Println("\nCount: ", counter)
}
```

## ATOMIC

```go
import (
	"fmt"
	"runtime"
	"sync"
	"time"
	"sync/atomic"
)

func main() {
	fmt.Println("CPUs: ", runtime.NumCPU())
	fmt.Println("Goroutines: ", runtime.NumGoroutine())

	var counter int64

	var wg sync.WaitGroup

	const gs = 100
	wg.Add(gs)

	fmt.Println()
	fmt.Println("Goroutines - Counter")
	for i := 0; i < gs; i++ {
		go func() {
			atomic.AddInt64(&counter, 1)
			runtime.Gosched()
			wg.Done()
			fmt.Printf("%v-%v\n", runtime.NumGoroutine(), atomic.LoadInt64(&counter))
		}()
	}
	fmt.Println()

	wg.Wait()

	fmt.Println("CPUs: ", runtime.NumCPU())
	fmt.Println("Goroutines: ", runtime.NumGoroutine())
	fmt.Println("\nCount: ", counter)
}
```

```go
import (
	"fmt"
)

func incrementor(s string) chan int {
	chInInt := make(chan int)

	go func(){
		for i := 0; i < 20; i++ {
			chInInt <- 1
			fmt.Printf("%v:%v ", s, i)
		}
		close(chInInt)
	}()

	return chInInt
}

func puller(chInInt chan int) chan int {
	chOutInt := make(chan int)

	go func(){
		var sum int
		for n := range chInInt {
			sum += n
		}
		chOutInt <- sum
		close(chOutInt)
	}()

	return chOutInt
}

func main() {
	chInInt1 := incrementor("Foo")
	chInInt2 := incrementor("Bar")
	chOutInt1 := puller(chInInt1)
	chOutInt2 := puller(chInInt2)
	fmt.Println("Final Counter: ", <-chOutInt1 + <-chOutInt2)
}
```

```go
import (
	"fmt"
)

func factorial(n int) chan int {
	chInt := make(chan int)
	go func(){
		total := 1
		for i := n; i > 0; i-- {
			total *= i
		}
		chInt <- total
		close(chInt)
	}()
	return chInt
}

func main() {
	chInt := factorial(4)
	for n := range chInt {
		fmt.Println(n)
	}
}
```

#### REDUX

```go
import (
	"fmt"
)

func gen() <-chan int {
	chInInt := make(chan int)

	go func(){
		for i := 0; i < 10; i++ {
			for j := 3; j < 13; j++ {
				chInInt <- j
			}
		}
		close(chInInt)
	}()

	return chInInt
}

func factorial(chInInt <-chan int) <-chan int {
	chOutInt := make(chan int)

	go func(){
		for n := range chInInt {
			chOutInt <- fact(n)
		}
		close(chOutInt)
	}()

	return chOutInt
}

func fact(n int) int {
	total := 1
	for i := n; i > 0; i-- {
		total *= i
	}
	return total
}

func main() {
	chInInt := gen()
	chOutInt := factorial(chInInt)

	for n := range chOutInt {
		fmt.Printf("%v ", n)
	}
}
```

### FAN IN

```go
import (
	"fmt"
	"sync"
)

func send(evenCh, oddCh chan<- int, wg *sync.WaitGroup) {
	defer close(evenCh)
	defer close(oddCh)
	defer wg.Done()

	for i := 0; i < 100; i++ {
		if i%2 == 0 {
			evenCh <- i
		} else {
			oddCh <- i
		}
	}
}

func receive(evenCh, oddCh <-chan int, fanin chan<- int, wg *sync.WaitGroup) {
	defer close(fanin)
	defer wg.Done()

	var faninWG sync.WaitGroup
	faninWG.Add(2)

	go func() {
		defer faninWG.Done()
		for v := range evenCh {
			fanin <- v
		}
	}()

	go func() {
		defer faninWG.Done()
		for v := range oddCh {
			fanin <- v
		}
	}()

	faninWG.Wait()
}

func main() {
	evenCh := make(chan int)
	oddCh := make(chan int)
	fanin := make(chan int)

	var wg sync.WaitGroup
	wg.Add(2)

	go send(evenCh, oddCh, &wg)
	go receive(evenCh, oddCh, fanin, &wg)

	go func() {
		wg.Wait()
		close(fanin)
	}()

	for v := range fanin {
		fmt.Printf("%v ", v)
	}

	fmt.Print("Exit ...")
}
```

```go
import (
	"fmt"
	"math/rand"
	"time"
)

func boring(msg string) <-chan string {
	chStr := make(chan string)

	go func(){
		for i := 0; ; i++ {
			chStr <- fmt.Sprintf("%v %v", msg, i)
			time.Sleep(time.Duration(rand.Intn(1e3))*time.Millisecond)
		}
		// close(chStr)
	}()

	return chStr
}

func fanIn(chStr1, chStr2 <-chan string) <-chan string {
	chStr := make(chan string)

	go func(){
		for {
			chStr <- <-chStr1
		}
	}()

	go func(){
		for {
			chStr <- <-chStr2
		}
	}()

	return chStr
}

func main() {
	chStr := fanIn(boring("Joe"), boring("Ann"))

	for i := 0; i < 10; i++ {
		fmt.Printf("%v ", <-chStr)
	}
	
	fmt.Println("\nYou're both boring. I'm leaving.")
}
```

### FAN OUT

```go
import (
	"fmt"
	"math/rand"
	"sync"
	"tim`e"
)

func populate(c chan int) {
	for i := 0; i < 100; i++ {
		c <- i
	}
	close(c)
}

func timeConsumingWork(n int) int {
	time.Sleep(time.Microsecond*time.Duration(rand.Intn(500)))
	return n + rand.Intn(1000)
}

func fanOutIn(c1, c2 chan int) {
	var wg sync.WaitGroup
	
	for v := range c1 {
		wg.Add(1)

		go func(v2 int) {
			c2 <- timeConsumingWork(v2)
			wg.Done()
		}(v)
	}

	wg.Wait()
	close(c2)
}

func main() {
	c1 := make(chan int)
	c2 := make(chan int)

	go populate(c1)

	go fanOutIn(c1, c2)

	for v := range c2 {
		fmt.Printf("%v ", v)
	}

	fmt.Println("\nExit ...")
}
```

```go
import (
	"fmt"
	"sync"
)

func gen(nums ...int) chan int {
	chInt := make(chan int)

	go func(){
		for _, v := range nums {
			chInt <- v
		}
		close(chInt)
	}()

	return chInt
}

func square(chInt chan int) chan int {
	chIntSquared := make(chan int)

	go func(){
		for n := range chInt {
			chIntSquared <- n*n
		}
		close(chIntSquared)
	}()

	return chIntSquared
}

func merge(chans ...chan int) chan int {
	chMerged := make(chan int)
	var wg sync.WaitGroup

	wg.Add(len(chans))

	for _, ch := range chans {
		go func(ch chan int) {
			defer wg.Done()
			for n := range ch {
				chMerged <- n
			}
		}(ch)
	}

	go func() {
		wg.Wait()
		close(chMerged)
	}()

	return chMerged
}

func main() {
	chInt := gen(2, 3)

	// Fan out
	chIntSquared1 := square(chInt) // worker 1
	chIntSquared2 := square(chInt) // worker 2

	// Fan in
	chMerged := merge(chIntSquared1, chIntSquared2)

	for n := range chMerged {
		fmt.Printf("%v ", n)
	} 
}
```

```go
import (
	"fmt"
	"math/rand"
	"sync"
	"time"
)

func populate(c chan int) {
	for i := 0; i < 100; i++ {
		c <- i
	}
	close(c)
}

func timeConsumingWork(n int) int {
	time.Sleep(time.Microsecond*time.Duration(rand.Intn(500)))
	return n + rand.Intn(1000)
}

func fanOutIn(c1, c2 chan int) {
	var wg sync.WaitGroup
	const goRoutines = 10
	wg.Add(goRoutines)

	for i := 0; i < goRoutines; i++ {
		go func() {
			for v := range c1 {
				func(v2 int) {
					c2 <- timeConsumingWork(v2)
					}(v)
				}
				
				wg.Done()
				}()
			}

	wg.Wait()
	close(c2)
}

func main() {
	c1 := make(chan int)
	c2 := make(chan int)

	go populate(c1)

	go fanOutIn(c1, c2)

	for v := range c2 {
		fmt.Printf("%v ", v)
	}

	fmt.Println("\nExit ...")
}
```

```go
import (
	"fmt"
)

func gen() <-chan int {
	chInInt := make(chan int)

	go func(){
		for i := 0; i < 100; i++ {
			for j := 3; j < 13; j++ {
				chInInt <- j
			}
		}
		close(chInInt)
	}()

	return chInInt
}

func factorial(chInInt <-chan int) <-chan int {
	chFactInt := make(chan int)

	go func(){
		for n := range chInInt {
			chFactInt <- fact(n)
		}
		close(chFactInt)
	}()

	return chFactInt
}

func fact(n int) int {
	total := 1
	for i := n; i > 0; i-- {
		total *= i
	}
	return total
}

func merge(chans ...<-chan int) <-chan int {
	var wg sync.WaitGroup
	chMerge := make(chan int)

	goRoutinePutValue := func(ch <-chan int) {
		for n := range ch {
			chMerge <- n
		}
		wg.Done()
	}

	wg.Add(len(chans))
	for _, ch := range chans {
		go goRoutinePutValue(ch)
	}

	go func(){
		wg.Wait()
		close(chMerge)
	}()

	return chMerge
}

func main() {
	chInInt := gen()

	chFactInt0 := factorial(chInInt)
	chFactInt1 := factorial(chInInt)
	chFactInt2 := factorial(chInInt)
	chFactInt3 := factorial(chInInt)
	chFactInt4 := factorial(chInInt)
	chFactInt5 := factorial(chInInt)
	chFactInt6 := factorial(chInInt)
	chFactInt7 := factorial(chInInt)
	chFactInt8 := factorial(chInInt)
	chFactInt9 := factorial(chInInt)

	chMerge := merge(chFactInt0, chFactInt1, chFactInt2, chFactInt3, chFactInt4,
									 chFactInt5, chFactInt6, chFactInt7, chFactInt8, chFactInt9)

	for n := range chMerge {
		fmt.Printf("%v ", n)
	}
}
```

## SELECT

```go


import "fmt"

func send(evenCh, oddCh, quitCh chan<- int) {
	for i := 0; i < 100; i++ {
		if i%2 == 0 {
			evenCh <- i
		} else {
			oddCh <- i
		}
	}

	close(evenCh)
	close(oddCh)
	close(quitCh)
}

func receive(evenCh, oddCh, quitCh <-chan int) {
	for {
		select {
		case v, ok := <-evenCh:
			if !ok {
				fmt.Println("Even Channel closed")
				evenCh = nil
			} else {
				fmt.Printf("Even Channel: %v\n", v)
			}
		case v, ok := <-oddCh:
			if !ok {
				fmt.Println("Odd Channel closed")
				oddCh = nil
			} else {
				fmt.Printf("Odd Channel: %v\n", v)
			}
		case v, ok := <-quitCh:
			if !ok {
				fmt.Println("Quit Channel closed")
				return
			}
			fmt.Printf("Quit Channel: %v\n", v)
		}
	}
}

func main() {
	evenCh := make(chan int)
	oddCh := make(chan int)
	quitCh := make(chan int)

	go send(evenCh, oddCh, quitCh)

	receive(evenCh, oddCh, quitCh)

	fmt.Print("Exit ...")
}
```

```go
import (
	"fmt"
)

func gen(q chan<- int) <-chan int {
	c := make(chan int)

	go func() {
		for i := 0; i < 100; i++ {
			c <- i
		}

		q <- 1

		close(c)
	}()

	return c
}

func receive(c, q <-chan int) {
	for {
		select {
		case v := <-c:
			fmt.Printf("%v ", v)
		case <-q:
			return
		}
	}
}

func main() {
	q := make(chan int)
	c := gen(q)

	receive(c, q)
}
```

```go
import (
	"fmt"
	"context"
)

func main() {
	ctx := context.Background()

	fmt.Printf("context:\t%v\n", ctx)
	fmt.Printf("context err:\t%v\n", ctx.Err())
	fmt.Printf("context type:\t%T\n", ctx)
	fmt.Println("----------")

	ctx, cancel := context.WithCancel(ctx)
	fmt.Printf("context:\t%v\n", ctx)
	fmt.Printf("context err:\t%v\n", ctx.Err())
	fmt.Printf("context type:\t%T\n", ctx)
	fmt.Printf("cancel:\t\t%v\n", cancel)
	fmt.Printf("cancel type:\t%T\n", cancel)
	fmt.Println("----------")

	cancel()
	fmt.Printf("context:\t%v\n", ctx)
	fmt.Printf("context err:\t%v\n", ctx.Err())
	fmt.Printf("context type:\t%T\n", ctx)
	fmt.Printf("cancel:\t\t%v\n", cancel)
	fmt.Printf("cancel type:\t%T\n", cancel)
	fmt.Println("----------")
}
```

```go
import (
	"context"
	"fmt"
	"runtime"
	"time"
)

func main() {
	ctx, cancel := context.WithCancel(context.Background())

	fmt.Println("\nError check 1: ", ctx.Err())
	fmt.Println("Num Goroutines 1: ", runtime.NumGoroutine(), "\n")

	go func() {
		n := 0

		for {
			select {
			case <-ctx.Done():
				return
			default:
				n++
				time.Sleep(time.Millisecond*200)
				fmt.Println("Working ", n)
			}
		}
	}()

	time.Sleep(time.Second*2)
	fmt.Println("\nError check 2: ", ctx.Err())
	fmt.Println("Num Goroutines 2: ", runtime.NumGoroutine(), "\n")

	fmt.Println("Cancel context ...")
	cancel()
	fmt.Println("Canceled context")

	time.Sleep(time.Second*2)
	fmt.Println("\nError check 3: ", ctx.Err())
	fmt.Println("Num Goroutines 3: ", runtime.NumGoroutine(), "\n")
}
```

## RECOVER

```go
import (
	"fmt"
)

func f() {
	defer func() {
		if r := recover(); r != nil {
			fmt.Println("Recovered in f", r)
		}
	}()

	fmt.Println("Calling g.")
	g(0)
	fmt.Println("Returned normally from g.")
}

func g(i int) {
	if i > 3 {
		fmt.Println("Panicking!")
		panic(fmt.Sprintf("%v", i))
	}
	defer fmt.Println("Defer in g", i)
	fmt.Println("Printing in g", i)
	g(i + 1)
}

func main() {
	f()
	fmt.Println("Returned normally from f.")
}
```

```go
import (
	"fmt"
	"log"
)

type norgateMathError struct {
	lat string
	long string
	err error
}

func (n norgateMathError) Error() string {
	return fmt.Sprintf("a norgate math error occured: %v %v %v", n.lat, n.long, n.err)
}

func sqrt(f float64) (float64, error) {
	if f < 0 {
		nme := fmt.Errorf("norgate math: square root of negative number (%v)", f)
		return 0, norgateMathError{"50.2289 N", "99.4656 W", nme}
	}
	return 42, nil
}

func main() {
	_, err := sqrt(-10)
	if err != nil {
		log.Println(err)
	}
}
```
